```python
df1=df
```

```python
max_thresold=df['price'].mean()+3*df['price'].std()
min_thresold=df['price'].mean()-3*df['price'].std()
```

```python
print("Max thresold is :",max_thresold)
print("Min thresold is :",min_thresold)
```

```python
df2 = df1[(df1['price'] > min_thresold) & (df1['price'] < max_thresold)]
df2.reset_index(drop=True, inplace=True)
```

One hot encoding

```python
df3 = pd.get_dummies(df2, columns=['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus'], dtype=int)
df3.head(5)
```

Linear Regression


```python
x=df3.drop(['price'],axis=1)
y=df3['price']
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_hat=model.predict(x_test)
```

```python
mst=mean_squared_error(y_hat,y_test)
r_sq=r2_score(y_hat,y_test)
```
